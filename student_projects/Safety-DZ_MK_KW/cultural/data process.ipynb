```python
import ast
import pandas as pd
import numpy as np

col_last = df.columns[-1]
col_prev = df.columns[-2]

def to_list(x):
    if isinstance(x, list):
        return x
    if pd.isna(x) or x == '[]':
        return []
    if isinstance(x, str):
        return ast.literal_eval(x)
    return []

df[col_last] = df[col_last].apply(to_list)
df[col_prev] = df[col_prev].apply(to_list)

mask = (
    df.index >= 71
    & df[col_last].apply(len).gt(0)
)

df.loc[mask, col_prev] = (
    df.loc[mask, col_prev] + df.loc[mask, col_last]
)
```

```python
df.to_csv(
    'popr.csv',
    encoding='utf-8-sig',
    sep=';',
    index=False
)
```
