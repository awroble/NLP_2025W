```python
import json
from pathlib import Path
```

### Creating file .json with prompts for LLM judge depending on the category, the .json file contains the .txt files as string values.

```python
bf_text = Path("BF.txt").read_text(encoding="utf-8")
jb_text = Path("JB.txt").read_text(encoding="utf-8")
mt_text = Path("MT.txt").read_text(encoding="utf-8")

bf_mt_text = Path("BF_MT.txt").read_text(encoding="utf-8")
jb_mt_text = Path("MT.txt").read_text(encoding="utf-8")


prefix_prompts = {
    "BF": bf_text,
    "MT": mt_text,
    "JB": jb_text,
    "JB_MT": jb_mt_text,
    "BF_MT": bf_mt_text
}


with open("judge_prompt.json", "w", encoding="utf-8") as f:
    json.dump(prefix_prompts, f, ensure_ascii=False, indent=2)
```
