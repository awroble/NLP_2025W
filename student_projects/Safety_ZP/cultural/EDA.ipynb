```python
import pandas as pd
from collections import Counter
import re

# Funkcja do bezpiecznego parsowania stringów
def parse_country_string(s):
    # Usuń nawiasy kwadratowe
    s = s.strip('[]')
    
    # Jeśli string jest pusty, zwróć pustą listę
    if not s:
        return []
    
    # Podziel po przecinkach, obsługując różne formaty cudzysłowów
    # Używamy wyrażenia regularnego, aby podzielić po przecinkach nie wewnątrz cudzysłowów
    parts = re.split(r',\s*(?=(?:[^"]*"[^"]*")*[^"]*$)', s)
    
    # Oczyść każdy element: usuń cudzysłowy i białe znaki
    countries = []
    for part in parts:
        # Usuń cudzysłowy (zarówno pojedyncze jak i podwójne)
        country = part.strip().strip('"\'').strip()
        if country:
            countries.append(country)
    
    return countries


ctr = pd.DataFrame()
df['parsed_countries'] = df['country'].apply(parse_country_string)
df_pic['parsed_countries'] = df_pic['country'].apply(parse_country_string)

all_countries = []
for country_list in df['parsed_countries']:
    all_countries.extend(country_list)
for country_list in df_pic['parsed_countries']:
    all_countries.extend(country_list)
    
country_counts = Counter(all_countries)

result_df = pd.DataFrame.from_dict(country_counts, orient='index', columns=['count']).sort_values('count', ascending=False)
result_df = result_df.reset_index()
result_df.columns = ['country','count']
```

```python

non_countries = {
    "Africa", "Latin America", "South Ossetia", "USSR", "Yugoslavia",
    "Palestine", 'Europa', 'Yoruba'
}
result_df = result_df[~result_df['country'].isin(non_countries)]
```

```python
# Dodaj nową kolumnę z liczbą słów dla każdego pytania
df['word_count'] = df['question'].str.split().str.len()

df_pic['word_count'] = df_pic['question'].str.split().str.len()


words = pd.concat([df['word_count'],df_pic['word_count']])
```

```python
plt.hist(words, bins = 20, color = 'green',)
plt.title('Number of words per question')
plt.xlabel('Number of words')
plt.ylabel('Number of occurances')
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Pobierz dane NLTK (tylko pierwszy raz)
nltk.download('punkt')

# Funkcja do liczenia tokenów
def count_tokens_nltk(text):
    if pd.isna(text):
        return 0
    tokens = word_tokenize(str(text))
    return len(tokens)

# Dodaj kolumnę z liczbą tokenów
df['nltk_token_count'] = df['question'].apply(count_tokens_nltk)
df_pic['nltk_token_count'] = df_pic['question'].str.split().str.len()

tokens = pd.concat([df['nltk_token_count'],df_pic['nltk_token_count']])



```

```python
word_tokens = pd.concat([words,tokens],axis = 1)
```

```python
plt.hist(word_tokens['nltk_token_count']-word_tokens['word_count'], bins = 20, color = 'green',)
plt.title('Difference between number of tokens and number of words')
plt.xlabel('Number of tokens minus number of words')
plt.ylabel('Number of occurances')
plt.show()
```
