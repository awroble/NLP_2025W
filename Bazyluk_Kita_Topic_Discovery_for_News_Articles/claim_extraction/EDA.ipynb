```python
import spacy
from graphviz import Digraph
nlp = spacy.load("en_core_web_trf")
```

```python
text = "US President Donald Trump's overseas envoy will travel to Germany this weekend to meet Ukrainian President Volodymyr Zelensky and European leaders for more talks on ending the war."
doc = nlp(text)
```

```python
def explore(doc):

    print("\n=== TOKENS ===")
    for token in doc:
        print(f"{token.text:15} | lemma: {token.lemma_:10} | pos: {token.pos_:6} | tag: {token.tag_:6} | dep: {token.dep_:10} | head: {token.head.text}")

    print("\n=== DEPENDENCY TREE ===")
    for token in doc:
        print(f"{token.text:15} --> {token.dep_:12} --> {token.head.text}")

    print("\n=== NAMED ENTITIES ===")
    for ent in doc.ents:
        print(f"{ent.text:20} | label: {ent.label_} | explanation: {spacy.explain(ent.label_)}")

    print("\n=== NOUN CHUNKS ===")
    for chunk in doc.noun_chunks:
        print(chunk)

    print("\n=== SENTENCE BOUNDARIES ===")
    for sent in doc.sents:
        print(sent)
```

```python
doc = nlp("He said in a video message that he would have preferred to see the process to the end, but national interest \"demanded otherwise\"")
explore(doc)
```

```python
from spacy import displacy
displacy.render(doc, style="dep", jupyter=True, options={"distance":   100})
```

```python
# WORKS BETTER IN COLAB

# Create a Graphviz Digraph
dot = Digraph(comment="Dependency Tree", format="png")
dot.attr(rankdir="TB")  # Top to Bottom layout
dot.attr('node', shape='ellipse', style='filled', fillcolor='lightgrey', fontname='Helvetica', fontsize='12')
dot.attr('edge', fontname='Helvetica', fontsize='10', color='black')

# Add nodes and edges
for token in doc:
    # Node: word\nPOS
    node_label = f"{token.text}\n({token.dep_})"
    dot.node(str(token.i), node_label)

    # Edge: head -> token
    if token.head.i != token.i:  # skip root self-loop
        dot.edge(str(token.head.i), str(token.i))

# Save and render
dot.render("spacy_tree.png", view=True)
```
