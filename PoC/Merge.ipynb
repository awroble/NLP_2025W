```python
import pandas as pd

imdb_df = pd.read_csv("data/imdb_processed.csv")
goodreads_df = pd.read_csv("data/goodreads_processed.csv")

imdb_cols = ["review_text", "is_spoiler", "content_words"]
goodreads_cols = ["review_text", "has_spoiler", "content_words"]

df_imdb_subset = imdb_df[imdb_cols].copy()

df_gr_subset = goodreads_df[goodreads_cols].copy()
df_gr_subset = df_gr_subset.rename(columns={"has_spoiler": "is_spoiler"})

df_imdb_subset['source'] = 'imdb'
df_gr_subset['source'] = 'goodreads'

combined_df = pd.concat([df_imdb_subset, df_gr_subset], axis=0, ignore_index=True)

print(f"IMDB shape: {df_imdb_subset.shape}")
print(f"Goodreads shape: {df_gr_subset.shape}")
print(f"Combined shape: {combined_df.shape}")
print("\nFirst 5 rows:")
print(combined_df.head())
```
